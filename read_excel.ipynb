```python
import matplotlib.pyplot as plt
import numpy as np

tasks = ["Power Up", "Integration", "Migration"]
months = ["June", "July", "August"]

# Convert month names to numerical values
month_map = {name: num for num, name in enumerate(months)}

plan = [month_map["July"], month_map["June"], month_map["August"]]
actual = [month_map["June"], month_map["July"], month_map["July"]]

bar_width = 0.35
index = np.arange(len(tasks))

fig, ax = plt.subplots()
bar1 = ax.bar(index, plan, bar_width, label='Plan')
bar2 = ax.bar(index + bar_width, actual, bar_width, label='Actual')

ax.set_xlabel('Tasks')
ax.set_ylabel('Months')
ax.set_title('Planned vs Actual Dates')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(tasks)
```

```python
import datetime

def transform_power_up(value):
    if pd.isnull(value) or value in ['', ' ', 'N/A']:
        return 'Missing Info'
    elif isinstance(value, datetime.datetime) or value in ['Y', 'N']:
        return 'Exists Info'
    else:
        return value

df_copy['Power Up '] = df_copy['Power Up '].apply(transform_power_up)


df_copy
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

value_counts = df_copy['Power Up '].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=value_counts.index, y=value_counts.values)
plt.xlabel('Values')
plt.ylabel('Count')
plt.title('Count of Values in Power Up')
plt.xticks(rotation=45)
plt.show()
```

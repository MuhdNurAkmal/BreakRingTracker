```python
import numpy as np
import pandas as pd
```

```python
df_dwdm = pd.read_excel("data/20240530_DWDM IP-BreakRing Tracker_URSZ (Batch 1) (3).xlsx", "Master Data", skiprows=[0])
df_dwdm.head(5)
```

```python
df_ipran = pd.read_excel("data/IPRAN YR23 YR24 Rollout Update - 06042024 - v1.1.xlsx", skiprows=[0])
df_ipran.head(5)
```

```python
print('Number of column in DWDM = ', len(df_dwdm.columns))
print('Number of column in IPRAN  = ', len(df_ipran.columns))
```

```python
df_dwdm.columns
```

```python
new_dwdm = df_dwdm[['Ring ID', 'DUID2', 'Batch', 'Priority Rollout (colour)',
                    'Site ID', 'Site Name', 'Subcon', 'CME Type', 'Status',
                    'Plan Survey Date', 'Survey Date', 'Plan MOS', 'Material On Site (MOS)',
                    'Plan Power Up', 'Power Up ', 'Plan Integration', 'Integration',
                    'Integrated Date (30days + install)', 'Break Ring Status'
                    ]]

new_dwdm.rename(columns = {'DUID2':'DUID / Activity Name'}, 
                inplace = True)
new_dwdm.info()
```

```python
df_ipran.columns
```

```python
new_ipran = df_ipran[['Ring ID', 'ACTIVITY NAME', 'Site ID', 'Site Name',
                      'Region', 'State', 'Subcon', 'Survey Plan',  'Survey Completed',
                      'SWAP/MIGRATION PLAN DATE', 'SWAP/MIGRATION \nCOMPLETION DATE',
                      'CD ES PO Date', 'CD EQ PO Date'
                      ]]
new_ipran.rename(columns = {'ACTIVITY NAME':'DUID / Activity Name'}, 
                inplace = True)
new_ipran.info()
```

```python
merged_df = pd.concat([new_dwdm, new_ipran])
merged_df.fillna(value='N/A', inplace=True)
merged_df.columns
```

```python
merged_df.rename(columns = {'Priority Rollout (colour)' : 'Priority',
                            'Status' : 'CME Status',
                            'Plan Survey Date' : 'Plan Survey',
                            'Survey Date' : 'Actual Survey',
                            'Material On Site (MOS)' : 'Actual MOS',
                            'Power Up ' : 'Actual Power Up',
                            'Integration' : 'Actual Integration',
                            'SWAP/MIGRATION PLAN DATE' : 'Plan Migration',
                            'SWAP/MIGRATION \nCOMPLETION DATE' : 'Actual Migration',
                            'CD ES PO Date' : 'PO ES Date', 
                            'CD EQ PO Date' : 'PO EQ Date'
                            }, 
                inplace = True)
```
